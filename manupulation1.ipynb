```python
%matplotlib inline
import imageio
import numpy as np
import matplotlib.pyplot as plt
pic=imageio.imread('/content/drive/MyDrive/Colab Notebooks/DataSet/p1.jpg')
gray=lambda rgb:np.dot(rgb[...,:3],[0.299,0.587,0.114])
gray=gray(pic)
max_=np.max(gray)
def log_transform():
  return(255/np.log(1+max_))*np.log(1+gray)
plt.figure(figsize=(5,5))
plt.imshow(log_transform(),cmap=plt.get_cmap(name='gray'))
plt.axis('off');

```

```python
import imageio
import matplotlib.pyplot as plt
#gama encoding
pic=imageio.imread('/content/drive/MyDrive/Colab Notebooks/DataSet/p1.jpg')
gamma=2.2 #gamma<1~Dark;gama>~Bright
gamma_correction=((pic/255)**(1/gamma))
plt.figure(figsize=(5,5))
plt.imshow(gamma_correction)
plt.axis('off');
```

```python
#image sharpness
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt
#load the image
my_image=Image.open('/content/drive/MyDrive/Colab Notebooks/DataSet/l1.jpg')
#use sharpen function
sharp=my_image.filter(ImageFilter.SHARPEN)
#save the image
#sharp.save('D:/image_sharpen.jpg')
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/DataSet/l1.jpg",0)
x,y=image.shape
z=np.zeros((x,y))
for i in range(0,x):
  for j in range(0,y):
    if (image[i][j]>50 and image[i][j]<150):
      z[i][j]-255
    else:
        z[i][j]=image[i][j]
equ=np.hstack((image,z))
plt.title('gray level slicing with background ')
plt.imshow(equ,'gray')
plt.show()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/DataSet/l1.jpg",0)
x,y=image.shape
z=np.zeros((x,y))
for i in range(0,x):
  for j in range(0,y):
      if (image[i][j]>50 and image[i][j]<150):
          z[i][j]=255
      else:
           z[i][j]=0
equ=np.hstack((image,z))
plt.title('gray level slicing without background ')
plt.imshow(equ,'gray')
plt.show()
```
